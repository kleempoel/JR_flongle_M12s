```r
library(tidyverse)
```

```r
setwd("C:/Data/Nanopore_JR/JR_Mimammal_testJM")
data=unique(read.table('blast/jr_flongle_m12s_blast_mito.txt',sep = '\t'))
colnames(data)=c('seqid', 'sseqid', 'sacc', 'staxids', 'evalue', 'pident', 'nident', 'slen', 'qstart', 'qend', 'length', 'mismatch')
```

```r
data[0:5,]
```

```r
eval_threshold=0.001
identity_threshold=85
print(paste("Number of records:",nrow(data)),quote = F)
print(paste("# records above evalue ",eval_threshold,':',length(which(data$evalue>eval_threshold)),
      ", below: ",length(which(data$evalue<eval_threshold))),quote = F)
print(paste("# records above identity ",identity_threshold,':',length(which(data$pident>identity_threshold)),
            ', below:',length(which(data$pident<identity_threshold))),quote = F)
data=data %>% filter(evalue<eval_threshold,pident>identity_threshold)
print(paste("blast records kept:",nrow(data)),quote = F)
```

```r
accessions=data %>% group_by(sacc) %>% summarise(count=n()) %>% arrange(-count)
print(nrow(accessions))
print(head(accessions,n = 10))
print(length(which(accessions$count>10)))
print(length(which(accessions$count>3)))
```
